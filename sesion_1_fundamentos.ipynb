```python
print('Esta es una calculadora')
print('Elije el tipo de operacion del menu')
print('1. Sumar')
print('2. Restar')
print('3. Multiplicar')
print('4. Dividir')
res=input('¿que quieres hacer? ')
num1=float(input('Escribe el primero numero: '))
num2=float(input('Escribe el primero numero: '))
if res=='Sumar':
  print('{} + {} = {}' .format (num1, num2, num1+num2,))
elif res=='Restar':
     print('{} - {} = {}' .format (num1, num2, num1-num2,))
elif res=='Multiplicar':
 print('{} * {} = {}' .format (num1, num2, num1*num2,))
elif res=='Dividir':
 print('{} / {} = {}' .format (num1, num2, num1/num2,))
else:
  print('Comando no valido')

```
